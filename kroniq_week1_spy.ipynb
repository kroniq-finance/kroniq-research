```python
import yfinance as yf
import matplotlib.pyplot as plt

# Pull 5 years of SPY data
spy = yf.download("SPY", start="2020-01-01", end="2025-12-31")

# Plot closing price
plt.figure(figsize=(12, 5))
spy["Close"].plot(
    title="SPY 5-Year Price Chart — Kroniq Research",
    color="#185FA5"
)
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.tight_layout()
plt.savefig("spy_5yr_chart.png", dpi=150)
plt.show()

print("Total rows of data:", len(spy))
print(spy.tail())
```
